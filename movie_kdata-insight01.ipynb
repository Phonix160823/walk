Step 1: Dataloading and overview of the datas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('tmdb-movies.csv')
df.head()

```

```python
# we check data types and number of missing rows
df.info()
```

Step 2: Clearning of the data. There is a to do list: a) Remove the missing and unnecessary columns/data. b)Remove the dublicate.

```python
#Delete the columns
df.drop(['id','imdb_id', 'homepage', 'tagline', 'keywords', 'runtime', 'overview', 'budget_adj', 'revenue_adj'], axis = 1, inplace = True)

df.info()
```

```python
#delite duplicates
df.drop_duplicates()
```

```python
# Delite the missing data
df[df.genres.isnull()]
df.dropna(inplace=True)
df.info()
```

```python
# Convert to rigth datatype

#convert revenue, budget from float to int
df.budget = df.budget.astype('int64')
df.revenue = df.revenue.astype('int64')

#Last check

df.info()
```

```python
# We look the datasummary
df.describe()
```

Step 3: Analyse of Data and knowledge acquisition

```python
# We use the insert function to create a new column named profit (located after revenue column)
df.insert(3,'profit',df['revenue']-df['budget'])
df.profit = df.profit.astype('int64')
# We show the first 2 rows to check the changes
df.head(2)
```

```python
df.profit.max()
```

The maximal Profit from a filmproduction is 2,544,505847 $, circa 2.54 Mlrd $.

```python
# We make a copie from the orginal
df_g = df

#First we have to seperate the genres, then count them, 
df_g.genres = df_g.genres.str.split('|')
df_g = df_g.explode('genres')

#make a new DataFrame 
genre = df_g['genres'].value_counts()
genre = pd.DataFrame(genre)
genre = genre.reset_index()
genre.rename({'index' : 'genre', 'genres' : 'count'}, axis = 1, inplace=True)
genre.sort_values('count', ascending=False, inplace=True)
#to check the DataFrame
genre.head(2)
```

```python
# We create with seaborn a view as barchart 
sns.barplot(x = genre['genre'], y=genre['count'],palette="PuOr_r")
plt.xticks(rotation=90)
plt.title('Number of films by genre', fontsize=16)
plt.show()
```

The favorit kind of genre is darma and comedy!

```python
# We want to know how is made the most movies
# To make is a copie from the orginal
df_p = df

# First we have to seperate the producer, then count them, 
df_p.production_companies = df_p.production_companies.str.split('|')
df_p = df_p.explode('production_companies')

#make a new DataFrame 
producer = df_p['production_companies'].value_counts()
producer = pd.DataFrame(producer)
producer = producer.reset_index()
producer.rename({'index' : 'producer', 'production_companies' : 'count'}, axis = 1, inplace=True)
producer.sort_values('count', ascending=False, inplace=True)
#to check the DataFrame
producer.head(5)
```

We see that "Universal Pictures" more films produced, the number of movies is 522

```python
# We group by release_year and then count them to see how much movies were released in each year.
df_yoy = df.groupby(['release_year'])['release_year'].count().reset_index(name='count')
#check the data Frame
df_yoy.head(2)
```

```python
# We show the most recent year and the most recent year
df_yoy.min(),df_yoy.max()
```

The record year by filmproduction is year 2015, the count of movies is 635

```python
# It woult be useful to visualize this

sns.lineplot(data=df_yoy, x='release_year', y='count', color='black')
plt.title('Development film production by year');
```

```python
# We array to dicreptived statistik
# We try to see a context or correlation using scatter plot
df.plot(x="popularity", y="vote_count", kind="scatter")
```

```python
# We want to see if there is a connection between popularity and time
df.groupby('popularity').release_year.mean().hist()
```
